### 行基本変形
- 「i行目をc倍する。」は、(i, i)成分がcである単位行列を左からかける。
- 「s行目にt行目のc倍を加える」は、(s, t)成分がcである単位行列を左からかける。
- 「p行目とq行目を入れ替える」は、単位行列のp行目の対角成分をq行目に、q行目の対角成分をo行目に変更する。

## 固有値

### 固有値と固有ベクトル

ある正方行列$A$に対して、以下のような式が成り立つような、特殊なベクトル$\vec{x}$と、右辺の係数$\lambda$がある。
$$A\vec{x} = \lambda \vec{x}$$

行列$A$とその特殊なベクトル$\vec{x}$の席は、ただのスカラーの数$\lambda$とその特殊なベクトル$\vec{x}$との積と同じ値になる。

この特殊なベクトル$\vec{x}$とその係数$\lambda$を、行列$A$に対する、固有ベクトル、固有値という。

### 固有値・固有ベクトル求め方

$A\vec{x} = \lambda\vec{x}$

$(A - \lambda I)\vec{x} = 0$

$\vec{x} \not= \vec{0}$より

$|A - \lambda I| = 0$
(行列$(A - \lambda I)$が逆行列を持たないということ)

### 固有値分解
ある実数を正方形にならべて作られた行列Aが固有値$\lambda_{1}$,$\lambda_{2}$,･･･と固有ベクトル$\vec{v_{1}}$, $\vec{v_{2}}$･･･を持ったとする。この固有値を対角線上に並べた行列（それ以外の成分は0）
$$\Lambda = \begin{pmatrix}
\lambda_{1} & & \\
& \lambda_{2} & \\
& & \ddots
\end{pmatrix}$$
と，それに対応する固有ベクトルを並べた行列
$$V = \begin{pmatrix}
\vec{v_{1}} & \vec{v_{2}}& \cdots\\
\end{pmatrix}$$
を用意したとき，それらは
$$A V = V \Lambda$$
と関係付けられる。したがって
$$A = V\Lambda V^{-1}$$
と変形できる。
このように正方形の行列を上述の様な３つの行列の積に変換することを固有値分解という。
この変換によって行列の累乗の計算が容易になる等の利点がある。

### 特異値
固有値・固有ベクトルの概念を一般の行列に拡張したもの。

### 特異値分解
$$M \vec v = \sigma \vec u$$
$$M^T \vec u = \sigma \vec v$$

$$M = USV^{-1}$$
- $\vec v: $ 単位ベクトル
- $\vec u: $ 単位ベクトル
- $U : m \times m$ の直交行列
- $V : n \times n$ の直交行列
- $S : m \times n$ の「非対角成分は0, 対角成分は非負で大きさの順に並んだ行列（下図のような行列）

### 特異値の求め方
$MV = US$
$M = USV^{-1}$

および、

$M^{T}V = VS^{T}$
$M^{T} = VS^{T}U^{-1}$

より、

$MM^{T} = USV^{-1}VS^{T}U^{-1} = USS^{T}U^{-1}$

つまり$MM^{T}$を固有値分解すれば，その左特異ベクトルと特異値の2乗が求められることがわかる。

## 特異値分解の利用例

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 画像を読み込み
A = cv2.imread('einstein.jpg')
print(A.shape)
# 白黒画像に変換
A = cv2.cvtColor(A, cv2.COLOR_BGR2GRAY)
print(A.shape)
```

```python
# 特異値分解
# A = usv^T
u, s, v = np.linalg.svd(A, full_matrices=False)
print(u.shape, s.shape, v.shape)

# 復元
B = np.dot(np.dot(u, np.diag(s)),v)
print((A.round(2) == B.round(2)).all())
```

```python
# オリジナル画像表示
plt.imshow(A, cmap="gray")
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(16, 10))

for i, k in enumerate([1, 2, 4, 8, 16, 32, 64, 128]):  
    
    ss = np.diag(s)[:k, :k]
    uu = u[:, :k]
    vv = v[:k, :]
    img_de = np.dot(np.dot(uu, ss), vv)
    
    ax = fig.add_subplot(2, 4, i+1)
    ax.axis("off")
    ax.set_title("k = {}".format(k), fontsize=20)
    plt.imshow(img_de, cmap="gray")
```

特異値分解にて画像解像度を下げる操作(次元削減)を行なった。
k = 4程度でも人間であることが推察でき、k = 16ではこの画像がアインシュタインであることが容易に理解できる。

### 確率
- 頻度確率(客観確率):ある事象が発生する頻度

- ベイズ確率(主観確率):信念の度合い

#### 確率の定義

$$P(A) = \frac{n(A)}{n(U)} = \frac{事象Aが起こる数}{すべての事象の数}$$

確率は0から1の値をとるので、$A$の補集合の確率$P(\bar A)$は、

$$P(\bar A) = \frac{n(\bar A)}{n(U)} = \frac{事象Aが起こらない数}{すべての事象の数}$$

$$= \frac{すべての事象の数 - 事象Aが起こる数}{すべての事象の数}= \frac{n(U) - n(A)}{n(U)} $$

$$= 1 - P(A)$$

AとBの共通部分$A \cap B$の確率$P(A \cap B)$は、
$$P(A \cap B) = P(A)P(B|A)$$
と表され、
$$A \cap B = B \cap A$$
より、
$$P(B \cap A) = P(B)P(A|B)$$
$$P(A)P(B|A) = P(B)P(A|B)$$

### 同時確率と条件付き確率
- 同時確率：全体の事象において、2つ以上の事象が同時に起こる確率

- 条件付き確率：ある事象の下で、事象が起こる確率

事象$X$と事象$Y$を考えるとき、
事象$X = x$が与えられた下で、事象$Y = y$となる条件付き確率は、
$$P(Y = y | X = x) = \frac{P(Y = y, X = x)}{P(X = x)}$$

事象$X$と事象$Y$が互いに独立な事象である場合、これら事象の同時確率は、
$$P(X = x, Y = y) = P(X = x)P(Y = y) = P(Y = y, X = x)$$

### ベイズの定理

先に示した条件付き確率の定義から導出される、

\begin{align}
P(Y = y | X = x) &= \frac{P(Y = y, X = x)}{P(X = x)} \\
&= \frac{P(X = x| Y = y)P(Y = y)}{P(X = x)}
\end{align}

をベイズの定理という。



#### 記述統計：集団の性質を要約し、記述する。

#### 推測統計：母集団から抽出された標本から母集団の性質を推測する。

#### 確率変数:事象と結び付けられた数値で、事象そのものを指すと解釈する場合も多い。

#### 確率分布：事象の発生する確率の分布。離散値であれば、表に示すことが可能。

### 期待値
分布における、確率変数のすべての値に確率の重みをつけた加重平均

事象：$X = x_{i}$

確率変数：$f(X) = f(x_{i})$

確率：$P(X) = P(x_{i})$

$i = 1, 2, \cdots, n$

のとき、

期待値$E(f)$は、

$$E(f) = \sum_{k = 1}^{n} P(X = x_{k})f(X = x_{k})$$

連続する値なら、
$$E(f) = \int P(X = x)f(X = x)dx$$

となる。

### 分散

データの散らばり具合のことで、データの各値が期待値からどれだけずれているかを平均したもの

分散$Var(f)$は、
\begin{align}
Var(f) &= E\bigl( (f(X = x) - E(f)) \bigr)\\
&= E(f^2(X = x) - (E(f))^2
\end{align}

### 共分散

2つのデータ系列の傾向の違い
正の値をとれば、似た傾向があり、負の値をとれば逆の傾向があることを示す。
ゼロであれば、関係性がないことを示す。

共分散$Cov(f, g)$は、
\begin{align}
Cov (f, g) &= E\Bigl( \bigl(f(X = x) - E(f)\bigr) \bigl(g(Y = y) - E(g)\bigr)\Bigr)\\
&= E(fg) - E(f)E(g)
\end{align}

### 標準偏差

分散の平方根

分散は2乗しているため元のデータと単位が異なる
→ 分散の平方根を求めることで、元の単位に戻る

標準偏差$\sigma$は、

$$\sigma = \sqrt {Var(f)}$$

### 様々な分布

#### ベルヌーイ分布
$$P(x|\mu) = \mu^x(1 - \mu)^{1 - x}$$

(例)コイントスなど

#### カテゴリカル分布
$$P(x|\mu) = \prod_{k = 1}^{K} \mu_{k} ^{x_k}$$

## 第3章：情報理論

### 自己情報量
自己情報量$I(x)$は、確率を$P(x)$、情報量を$W(x)$としたとき、
$$I(x) = - \log (P(x)) = \log (W(x))$$
と表す。

### シャノンエントロピー
シャノンエントロピー$H(x)$とは、自己情報量の期待値のことで、

\begin{align}
H(x) &= E\Bigl(I(x)\Bigr)\\
&= - E \Bigl(\log \bigl(P(x) \bigr)\Bigr)\\
&= - \sum \Bigl(P(x) \log \bigl( P(x) \bigr) \Bigr)
\end{align}

と表される。

### カルバック・ライブラー・ダイバージェンス

カルバック・ライブラーダイバージェンス$D_{KL}(P||Q)$とは、同じ事象、確率変数における異なる確率分布$P$、$Q$の違いを表す。厳密には距離ではない。

\begin{align}
D_{KL}(P||Q) &= \mathbb{E}_{x \sim p}\Bigl[\log \frac{P(x)}{Q(x)}\Bigr]\\
&= \mathbb{E}_{x \sim p} \Bigl[\log P(x) - \log Q(x) \Bigr]\\
&= \sum_{x} P(x) \log \frac{P(x)}{Q(x)}
\end{align}

ここで、$\log P(x) - \log Q(x) = I(Q(x)) - I(P(x))$であることに注意。

### 交差エントロピー

交差エントロピー$H(P, Q)$とは、KLダイバージェンスの一部を取り出したもので、$Q$についての自己情報量を$P$の分布で平均したものである。$D_{KL}(P||Q) = \sum_{x} P(x) \Bigl(-\log {Q(x)} - \log {P(x)} \Bigr)$の前半部分

\begin{align}
H(P, Q) &= H(P) + D_{KL}(P||Q) \\
&= \mathbb{E}_{x \sim p} \log {Q(x)}\\
&= - \sum_{x} P(x) \log {Q(x)}
\end{align}
